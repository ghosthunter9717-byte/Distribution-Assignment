4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

   - Standardization is a data transformation technique that rescales data to have a mean of zero and a standard deviation of one.

   - To calculate the z-score for a value in excel, the STANDARDIZE function can be used. Asuming the value is in cell A2, the mean is in cell B2, and the standard deviation is in cell C2 the formula is:

   =STANDARDIZE(A2,B2,C2)

5. What is Kurtosis and their type?

   - Kurtosis is a statistical measure that describe the "tailedness" of a distribution's probability distribution. The three main types are: Mesokurtic, Leptokurtic, and playkurtic.
